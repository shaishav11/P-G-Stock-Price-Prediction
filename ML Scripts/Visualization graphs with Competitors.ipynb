```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline


from scipy.optimize import minimize
from pandas_datareader import DataReader
from datetime import datetime
from sklearn.neighbors import KernelDensity

import seaborn as sns
```

```python
PG = DataReader('PG',  'yahoo', datetime(2000,1,3), datetime(2019,12,31));
JNJ=  DataReader('JNJ',  'yahoo', datetime(2000,1,3), datetime(2019,12,31));
UL=  DataReader('UL',  'yahoo', datetime(2000,1,3), datetime(2019,12,31));
CL=  DataReader('CL',  'yahoo', datetime(2000,1,3), datetime(2019,12,31));
CHD=  DataReader('CHD',  'yahoo', datetime(2000,1,3), datetime(2019,12,31));
```

```python
PG.head()
```

```python
PG['Open'].plot(label='P&G',figsize=(16,8),title='Opening Price')
JNJ['Open'].plot(label='Johnson&Johnson')
UL['Open'].plot(label='Unilever')
CL['Open'].plot(label='Colgate-Palmolive')
CHD['Open'].plot(label='Church & Dwight')
plt.legend()
```

```python
PG['High'].plot(label='P&G',figsize=(16,8),title='Highest Price')
JNJ['High'].plot(label='Johnson&Johnson')
UL['High'].plot(label='Unilever')
CL['High'].plot(label='Colgate-Palmolive')
CHD['High'].plot(label='Church & Dwight')
plt.legend()
```

```python
PG['MA50']=PG['Adj Close'].rolling(50).mean()
PG['MA200']=PG['Adj Close'].rolling(200).mean()
PG[['Adj Close','MA50','MA200']].plot(label='P&G',figsize=(16,8));
```
